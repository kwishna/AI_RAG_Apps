```python
import ollama
```

```python
response = ollama.chat(model='gemma:2b-instruct', messages=[
  {
    'role': 'user',
    'content': 'Why is the sky blue?',
  },
])
print(response)
```

```python
print(response['message']['content'])
```

```python
resp = ollama.generate(model='gemma:2b-instruct', prompt='Why is the sky blue?')
print(resp)
```

```python
print(resp['response'])
```

```python
emb = ollama.embeddings(model='gemma:2b-instruct', prompt='They sky is blue because of rayleigh scattering')
print(emb)
```

```python
print(len(emb['embedding']))
```

```python
stream = ollama.chat(
    model='gemma:2b-instruct',
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}],
    stream=True,
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```
